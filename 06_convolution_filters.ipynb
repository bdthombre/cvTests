```python
import scipy
import numpy as np
import cv2
import matplotlib.pyplot as plt
import scipy.ndimage
```

```python
img = cv2.imread("tests/test_image.jpg",0)
img.shape
```

```python
k_sharp = np.array([[-0.,-1.,-0],[-1.,5.,-1.],[-0.,-1.,0]])

k_sobel = np.array([[-1.,-2.,-1.],[-0.,0.,-0.],[1.,2.,1]])
k = np.array([[-1.,-2.,-1.],[-0.,0.,-0.],[1.,2.,1]])
k.shape, k
```

```python
plt.imshow(img, cmap="gray")
```

```python
convolve = scipy.ndimage.convolve(img, k)
```

```python
convolve.shape
```

```python
plt.imshow(convolve, cmap='gray')
```

```python
# sobel
sobel_x = np.array([[-1, 0, 1],
                    [-2, 0, 2],
                    [-1, 0, 1]])

sobel_y = np.array([[-1, -2, -1],
                    [0, 0, 0],
                    [1, 2, 1]])


gradient_x = scipy.ndimage.convolve(img, sobel_x)
gradient_y = scipy.ndimage.convolve(img, sobel_y)

gradient_magnitude = np.sqrt(gradient_x**2 + gradient_y**2)
plt.imshow(gradient_magnitude, cmap='gray')
```

```python
# with cv2
sobel = cv2.Sobel(img, cv2.CV_64F, 1,0, ksize=3)
plt.imshow(sobel, cmap='gray')
```

```python

# Apply Sobel operator in x direction
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)

# Apply Sobel operator in y direction
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)

# Compute the magnitude of the gradient
gradient_magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Convert the result to 8-bit for display
sobel_x = cv2.convertScaleAbs(sobel_x)
sobel_y = cv2.convertScaleAbs(sobel_y)
gradient_magnitude = cv2.convertScaleAbs(gradient_magnitude)

plt.imshow(gradient_magnitude, cmap='gray')
```
