```python
import cv2
import scipy
import numpy as np
import scipy.signal
import matplotlib.pyplot as plt
```

```python
file = "tests/paul_mcarty.mp4"
cap = cv2.VideoCapture(file)
```

```python
start_frame = 200
diff_frame = 5
```

```python
cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame)
ret, frame1 = cap.read()


cap.set(cv2.CAP_PROP_POS_FRAMES, start_frame+diff_frame)
ret, frame2 = cap.read()
```

```python
fig, (ax1, ax2)= plt.subplots(1,2)
ax1.imshow(frame1)
ax2.imshow(frame2)
```

```python
frame1 = cv2.cvtColor(frame1, cv2.COLOR_BGR2GRAY)
frame2 = cv2.cvtColor(frame2, cv2.COLOR_BGR2GRAY)
```

```python
combined = cv2.subtract(frame1, frame2)
plt.imshow(combined, cmap='gray')
```
